```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
medicare = pd.read_csv(r"C:\Users\User\OneDrive\Desktop\Cleaned_MediCare_Data GroupA(1).csv")
```

```python
medicare
```

```python
revenue_by_dept = medicare.groupby('department')['revenue_generated'].sum()
```

```python
medicare
```

```python
count_by_insurance = medicare['insurance_type'].value_counts()
```

```python
readmit_by_treatment = medicare.groupby('treatment_type')['readmission_rate_(%)'].mean().sort_values(ascending=False)
```

```python
profit_margin_by_doc = medicare.groupby('doctor_name')['profit_margin'].mean().sort_values(ascending=False)
```

```python
# Bar chart representative of Total Revenue by Department
revenue_by_dept.plot(kind='barh', figsize=(10,6), color='forestgreen')
plt.title("Total Revenue per Department")
plt.xlabel("Revenue")
plt.ylabel("Department")
plt.tight_layout()
plt.show()
```

#Revenue Insights  
The bar chart highlights that Emergency and Oncology departments generate the highest revenue, indicating their critical role in hospital operations.

```python
#Pie chart representation for Patients by Insurancs Type
count_by_insurance.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(6,6),colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title("Patient Distribution by Insurance Type")
plt.ylabel("")
plt.show()
```

#This pie chart illustrates the proportion of patients covered under different insurance types. Medicaid and Medicare account for the majority of insured patients, while private insurance represents a smaller yet significant share. The uninsured category, though lower, may indicate gaps in healthcare access.

```python
#Bar chat representative for Average Readmission rate by Treatment type
readmit_by_treatment.plot(kind='bar', figsize=(10,5), color='forestgreen')
plt.title("Average Readmission Rate by Treatment Type")
plt.xlabel("Treatment Type")
plt.ylabel("Readmission Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The heatmap identifies cases where patients received treatments not typically associated with their gender. Notably, mismatches appear in Maternity and certain surgical procedures. Addressing these inconsistencies could refine hospital policies and improve patient experiences.

```python
#A bar cht representation of Average Profit margin by Doctor
profit_margin_by_doc.plot(kind='barh', figsize=(10,6), color='forestgreen')
plt.title("Average Profit Margin by Doctor")
plt.xlabel("Profit Margin (%)")
plt.ylabel("Doctor")
plt.tight_layout()
plt.show()
```

The horizontal bar chart reveals variability in profit margins among doctors, with some achieving consistently high returns while others show losses. This disparity could be linked to treatment types, patient demographics, or operational efficiencies within different departments
